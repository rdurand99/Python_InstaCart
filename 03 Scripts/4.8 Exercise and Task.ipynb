# Step 6: 
The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: Low Spender and High Spender

```python
#Descriptive statisitcs for prices column
print('Descriptive statistics for prices columns')
ords_prods_merge['prices'].describe()
```

```python
#Creating avg price column using transform()
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#Lower spender flag
ords_prods_merge.loc[ords_prods_merge['average_price'] <= 10, 'spending_flag'] = 'Low spender'
```

```python
#Higher spender flag
ords_prods_merge.loc[ords_prods_merge['average_price'] > 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
#Checking the frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Step 7:
In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. 

```python
#Creating median for days since prior order column using transform()
ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#Non-frequent customer flag
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#Regular customer flag
ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] <= 20) & (ords_prods_merge['median_prior_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
#Frequent customer flag
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
#Checking the frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
