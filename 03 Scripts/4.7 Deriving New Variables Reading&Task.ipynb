### Creating if-statments

```python
#Setting up a user-defined function for price ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

#Step. 4 Identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day

```python
#Created dataframe for count of values for order_hour_of day
counts = ords_prods_merged['order_hour_of_day'].value_counts().rename_axis('hour').reset_index(name='counts')
```

```python
counts
```

#### Created a rename_axis() because I needed to rename the index of the resulting data frame to ‘hour’
#### Created reset_index() because I needed to reset the data frame into a column, specifically resetting it to a column called ‘index’ and renaming it ‘hour’

```python
#Created a list showing the percentiles for the value conts of ords_prods_merged
orders = list(ords_prods_merged['order_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
orders
```

#### Created quantiles to calculate the distribution of order counts by percentile 

Created loop for “Most orders,” “Average orders,” and “Fewest orders.”

```python
#Fewest Ords list
fewest_ords = list(counts.loc[counts['counts'] <= orders[0]]['hour'])
```

```python
#Average Ords List
avg_ords = list(counts[(counts['counts'] > orders[0]) & (counts['counts'] < orders[2])]['hour'])
```

```python
#Most Ords List
mos_ords = list(counts.loc[counts['counts'] >= orders[2]]['hour'])
```

Created indexes for “Most orders,” “Average orders,” and “Fewest orders.”

```python
#Created indexes for “Most orders,” “Average orders,” and “Fewest orders.”
#Fewest Ords Index
index_fewest_ords = ords_prods_merged['order_hour_of_day'].isin(fewest_ords)
```

```python
#Avg Ords Index
index_avg_ords = ords_prods_merged['order_hour_of_day'].isin(avg_ords)
```

```python
#Most Ords Index
index_mos_ords = ords_prods_merged['order_hour_of_day'].isin(mos_ords)
```

Create a new column containing these labels called “busiest_period_of_day

```python
ords_prods_merged.loc[index_fewest_ords,'busiest_period_of_day'] = 'Fewest orders'
```

```python
ords_prods_merged.loc[index_avg_ords,'busiest_period_of_day'] = 'Average orders'
```

```python
ords_prods_merged.loc[index_mos_ords,'busiest_period_of_day'] = 'Most orders'
```
